# Mental Health in Tech Survey

This dataset contains the following data:
- Timestamp
- Age
- Gender
- Country
- state: If you live in the United States, which state or territory do you live in?
- self_employed: Are you self-employed?
- family_history: Do you have a family history of mental illness?
- treatment: Have you sought treatment for a mental health condition?
- work_interfere: If you have a mental health condition, do you feel that it interferes with your work?
- no_employees: How many employees does your company or organization have?
- remote_work: Do you work remotely (outside of an office) at least 50% of the time?
- tech_company: Is your employer primarily a tech company/organization?
- benefits: Does your employer provide mental health benefits?
- care_options: Do you know the options for mental health care your employer provides?
- wellness_program: Has your employer ever discussed mental health as part of an employee wellness program?
- seek_help: Does your employer provide resources to learn more about mental health issues and how to seek help?
- anonymity: Is your anonymity protected if you choose to take advantage of mental health or substance abuse treatment resources?
- leave: How easy is it for you to take medical leave for a mental health condition?
- mental_health_consequence: Do you think that discussing a mental health issue with your employer would have negative consequences?
- phys_health_consequence: Do you think that discussing a physical health issue with your employer would have negative consequences?
- coworkers: Would you be willing to discuss a mental health issue with your coworkers?
- supervisor: Would you be willing to discuss a mental health issue with your direct supervisor(s)?
- mental_health_interview: Would you bring up a mental health issue with a potential employer in an interview?
- phys_health_interview: Would you bring up a physical health issue with a potential employer in an interview?
- mental_vs_physical: Do you feel that your employer takes mental health as seriously as physical health?
- obs_consequence: Have you heard of or observed negative consequences for coworkers with mental health conditions in your workplace?
- comments: Any additional notes or comments

2. Data Cleaning and Transformation

- step 1: Drop non necessary features  ('comments','Timestamp')
- step 2: Age:- filter out unreasonable value and fill with median.
- step 3: Gender:- recode into appriate form.
- step 4: state:- drop cause of high missing value
- step 5: self_employed:- fill missing value with mode
- step 6: work_inference:- fill missing value with unknown

```python
# rempve comments and Timestamp column
df.drop(columns=['comments', 'Timestamp'], inplace=True)
```

```python
# Filter out unreasonable ages assuming valid age is from 10 to 100.
df.loc[(df['Age'] < 10) | (df['Age'] > 100), 'Age'] = pd.NA

# fill missing value with median
df['Age']=df['Age'].fillna(df['Age'].median())
```

```python
df['Gender'].unique()
```

```python
# define a function to recode into appropriate form
def clean_gender(gender):
    gender = str(gender).strip().lower()
    if gender in ['male', 'm', 'cis male', 'man', 'mal', 'maile', 'msle', 'make', 'malr', 'cis man', 'male ', 'mail']:
        return 'Male'
    elif gender in ['female', 'f', 'cis female', 'woman', 'femake', 'femail', 'female ', 'female (cis)', 'female (trans)', 'cis-female/femme']:
        return 'Female'
    else:
        return 'Other'

# appy above function
df['Gender']=df['Gender'].apply(clean_gender)

# cross check 
df['Gender'].value_counts()
```

```python
# Drop 'state' due to high missing values
df.drop(columns=['state'], inplace=True)
```

```python
df['self_employed'].value_counts(dropna = False)
df['self_employed'] = df['self_employed'].fillna(df['self_employed'].mode()[0])
```

```python
df['work_interfere'].value_counts(dropna = False)
df['work_interfere'] = df['work_interfere'].fillna('unknown')
```

```python
# keep top 10 country as it is and recode remaning as 'other'. 
Top_10_country = list(df['Country'].value_counts().sort_values(ascending=False).head(10).index)
df['Country'] = df['Country'].apply(lambda x: x if x in Top_10_country else 'other')
df['Country'].value_counts()
```

Mann-Whitney U test: check whether two independent groups differ significantly in terms of a continuous variable 

```python
from scipy.stats import mannwhitneyu

group1 = df[df['treatment'] == 'Yes']['Age']
group2 = df[df['treatment'] == 'No']['Age']

stat, p = mannwhitneyu(group1, group2, alternative='two-sided')
print(f"Mann-Whitney U Test p-value: {p:.4f}")

if p < 0.05:
    print("Conclusion: Age distribution is significantly different between treatment groups.")
else:
    print("Conclusion: No significant difference in age distribution between treatment groups.")
```

conclusion: 
- Although the box plot suggests that age may not have a strong visual impact on whether someone received treatment, the Mann-Whitney U test reveals a statistically significant difference in age distribution between treatment groups (p-value: 0.0125). This indicates that age is likely associated with treatment status, even if the visual differences appear subtle.

Chi-square Test for Categorical Variables vs Treatment

```python
from scipy.stats import chi2_contingency

# Store p-values
chi2_results = {}

for col in cat_indep_columns:
    contingency_table = pd.crosstab(df[col], df['treatment'])
    
    if contingency_table.shape[0] > 1 and contingency_table.shape[1] > 1:  # Skip if no variation
        chi2, p, dof, ex = chi2_contingency(contingency_table)
        chi2_results[col] = round(p, 4)

# Convert to DataFrame
chi2_df = pd.DataFrame.from_dict(chi2_results, orient='index', columns=['p_value']).sort_values('p_value')
print(chi2_df)
```

```python
# spliting into trainning and testing data 
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=.2,random_state=42)
```

```python
cat_feature = [column for column in df if df[column].dtype == 'object']
for i in cat_feature:
    print(i,'\n',df[i].unique())
```

```python
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder, MinMaxScaler
from sklearn.compose import ColumnTransformer

# Define columns to be used in each transformer
ordinal_features = ['family_history', 'work_interfere', 'benefits', 'care_options', 'wellness_program',
                    'seek_help', 'anonymity', 'leave', 'mental_health_consequence', 'coworkers',
                    'mental_health_interview', 'mental_vs_physical', 'obs_consequence']

# Make sure the number of category lists matches the number of ordinal features
ordinal_categories = [
    ['No', 'Yes'],  # family_history
    ['unknown', 'Never', 'Rarely', 'Sometimes', 'Often'],  # work_interfere
    ['No', "Don't know", 'Yes'],  # benefits
    ['No', "Not sure", 'Yes'],  # care_options
    ['No', "Don't know", 'Yes'],  # wellness_program
    ['No', "Don't know", 'Yes'],  # seek_help
    ['No', "Don't know", 'Yes'],  # anonymity
    ['Very easy', 'Somewhat easy', 'Somewhat difficult', 'Very difficult', "Don't know"],  # leave
    ['No', 'Maybe', 'Yes'],  # mental_health_consequence
    ['No', 'Some of them', 'Yes'],  # coworkers
    ['No', 'Maybe', 'Yes'],  # mental_health_interview
    ['No', "Don't know", 'Yes'],  # mental_vs_physical
    ['No', 'Yes']  # obs_consequence
]

# ColumnTransformer with debugging fixes
preprocessor = ColumnTransformer(
    transformers=[
        ('ordered_encoding', OrdinalEncoder(categories=ordinal_categories), ordinal_features),
        ('onehotencoder', OneHotEncoder(drop='first'), ['Country', 'Gender']),
        ('minmax', MinMaxScaler(), ['Age'])
    ],
    remainder='drop'  # Drop remaining columns not specified
)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier
from sklearn.svm import SVC
#from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, precision_score, recall_score, f1_score, roc_auc_score

models = {
   # "GaussianNB":GaussianNB(), 
    "XGBClassifier": XGBClassifier(),
    "Logistic Regression" : LogisticRegression(max_iter=500),
    "Decision Tree" : DecisionTreeClassifier(),
    "Random Forest" : RandomForestClassifier(random_state=42),
    "svm" :SVC() 
}


for i in range(len(list(models))):
    model = Pipeline(steps=[
        ('preprocessing', preprocessor),
        (list(models.keys())[0], list(models.values())[i])
    ])
    
    model.fit(x_train, y_train) # Train the model

    # Make predictions
    y_train_pred = model.predict(x_train)
    y_test_pred = model.predict(x_test)

    # Training set performance
    mode_train_accuracy = accuracy_score(y_train, y_train_pred) # calculate accuracy
    model_train_f1 = f1_score(y_train, y_train_pred) # calculate f1 score
    model_train_precision = precision_score(y_train, y_train_pred) # calculate precision
    model_train_recall = recall_score(y_train, y_train_pred) # calculate recall
    model_train_roc_auc = roc_auc_score(y_train, y_train_pred) # calculate roc_auc

    # Test set performance
    model_test_accuracy = accuracy_score(y_test, y_test_pred) # calculate accuracy
    model_test_f1 = f1_score(y_test, y_test_pred) # calculate f1 score
    model_test_precision = precision_score(y_test, y_test_pred) # calculate precision
    model_test_recall = recall_score(y_test, y_test_pred) # calculate recall
    model_test_roc_auc = roc_auc_score(y_test, y_test_pred) # calculate roc_auc

    print(f"Model: {list(models.keys())[i]}")

    print("Training set performance")
    print("Accuracy: {:.4f}".format(mode_train_accuracy))
    print("F1 Score: {:.4f}".format(model_train_f1))
    print("Precision: {:.4f}".format(model_train_precision))
    print("Recall: {:.4f}".format(model_train_recall))
    print("ROC AUC: {:.4f}".format(model_train_roc_auc))

    print("\nTest set performance")
    print("Accuracy: {:.4f}".format(model_test_accuracy))
    print("F1 Score: {:.4f}".format(model_test_f1))
    print("Precision: {:.4f}".format(model_test_precision))
    print("Recall: {:.4f}".format(model_test_recall))
    print("ROC AUC: {:.4f}\n".format(model_test_roc_auc))
```

```python
# 1. Get the transformers
feature_names = []

for name, transformer, cols in preprocessor.transformers_:
    if name == 'remainder':
        continue
    if hasattr(transformer, 'get_feature_names_out'):
        # OneHotEncoder supports get_feature_names_out
        names = transformer.get_feature_names_out(cols)
    elif isinstance(transformer, OrdinalEncoder):
        # OrdinalEncoder doesn't rename columns
        names = cols
    elif isinstance(transformer, MinMaxScaler):
        # Same for MinMaxScaler
        names = cols
    else:
        # Fallback
        names = cols
    feature_names.extend(names)


x_train_df = pd.DataFrame(X_train_preprocessed, columns=feature_names, index=x_train.index)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

# 1. Sort features by importance descending
sorted_features = feat_imp_df['Feature'].tolist()

accuracies = []

# 2. Loop over top k features from 1 to all
for k in range(1, len(sorted_features) + 1):
    top_k_features = sorted_features[:k]

    # Select columns
    X_train_sel = x_train_df[top_k_features]
    X_test_sel = x_test_df[top_k_features]  # make sure you have test data in dataframe with same columns

    # Train model with best params
    model = RandomForestClassifier(
        n_estimators=200,
        max_depth=None,
        max_features='log2',
        min_samples_leaf=2,
        min_samples_split=2
    )
    model.fit(X_train_sel, y_train)

    # Predict on test
    y_pred = model.predict(X_test_sel)

    # Calculate accuracy
    acc = accuracy_score(y_test, y_pred)
    accuracies.append(acc)

    print(f"Top {k} features accuracy: {acc:.4f}")

# 3. Plot accuracy vs number of features
plt.figure(figsize=(10,6))
plt.plot(range(1, len(sorted_features) + 1), accuracies, marker='o')
plt.xlabel('Number of Top Features')
plt.ylabel('Test Accuracy')
plt.title('Test Accuracy vs Number of Important Features')
plt.grid(True)
plt.show()
```
